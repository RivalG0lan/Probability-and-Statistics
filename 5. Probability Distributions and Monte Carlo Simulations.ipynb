```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  
import seaborn as sns
from scipy.stats import norm, binom
import scipy.stats as stats
```

```python
dice_rolls = np.random.randint(1, 7, 10000)

plt.figure(figsize=(8, 4))
sns.histplot(dice_rolls, bins=np.arange(1, 8) - 0.5, kde=False)

plt.xticks(range(1, 7))
plt.title("Simulation of Rolling a Die 10,000 Times")
plt.xlabel("Die Face")
plt.ylabel("Frequency")
plt.show()
```

```python
# binomial distribution parameters
n = 5   #number of trials
p = 0.5 #probability of success (e.g., getting heads)
k = 3   #desired number of successes (e.g., 3 heads)

# calculate probability of exactly 3 successes
prob_3_heads = binom.pmf(k, n, p)

print(f"Probability of getting exactly 3 heads out of 5 coin tosses: {prob_3_heads:.4f}")
```

```python
np.random.seed(42)
normal_data = np.random.normal(loc=50, scale=18, size=1000)

sns.histplot(normal_data, bins=30, kde=True)
plt.title("Continuous Random Variable: Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
sims = 1000000
A = np.random.uniform(1, 5, sims)
B = np.random.uniform(2, 6, sims)
duration = A + B

plt.figure(figsize=(3, 1.5))
plt.hist(duration, density=True)
plt.axvline(9, color='r')
plt.show()

prob = (duration > 9).sum() / sims
print(f"Estimated probability of Duration > 9: {prob:.4f}")
```

```python
# Monte Carlo Simulation - 1st run
sims = 10_000_000
A = np.random.uniform(1, 5, sims)
B = np.random.uniform(2, 6, sims)
duration = A + B

plt.figure(figsize=(6, 3))
plt.hist(duration, bins=50, density=True, alpha=0.6, color='b')
plt.axvline(9, color='r', linestyle='dashed', linewidth=2)
plt.title("Monte Carlo Simulation: A + B")
plt.xlabel("Duration")
plt.ylabel("Density")
plt.show()

prob = (duration > 9).sum() / sims
print(f"Estimated probability of Duration > 9: {prob:.4f}")

# Monte Carlo Simulation - 2nd run (P(Duration < 9))
A = np.random.uniform(1, 5, sims)
B = np.random.uniform(2, 6, sims)
duration = A + B

plt.figure(figsize=(6, 3))
plt.hist(duration, bins=50, density=True, alpha=0.6, color='b')
plt.axvline(9, color='r', linestyle='dashed', linewidth=2)
plt.title("Monte Carlo Simulation: A + B")
plt.xlabel("Duration")
plt.ylabel("Density")
plt.show()

prob = (duration < 9).sum() / sims
print(f"Estimated probability of Duration < 9: {prob:.4f}")

# Monte Carlo Simulation - 3rd run (P(Duration > 9))
A = np.random.uniform(1, 5, sims)
B = np.random.uniform(2, 6, sims)
duration = A + B

plt.figure(figsize=(6, 3))
plt.hist(duration, bins=50, density=True, alpha=0.6, color='b')
plt.axvline(9, color='r', linestyle='dashed', linewidth=2)
plt.title("Monte Carlo Simulation: A + B")
plt.xlabel("Duration")
plt.ylabel("Density")
plt.show()

prob = (duration > 9).sum() / sims
print(f"Estimated probability of Duration > 9: {prob:.4f}")

# Monte Carlo Simulation - 4th run (P(Duration < 9))
A = np.random.uniform(1, 5, sims)
B = np.random.uniform(2, 6, sims)
duration = A + B

plt.figure(figsize=(6, 3))
plt.hist(duration, bins=50, density=True, alpha=0.6, color='b')
plt.axvline(9, color='r', linestyle='dashed', linewidth=2)
plt.title("Monte Carlo Simulation: A + B")
plt.xlabel("Duration")
plt.ylabel("Density")
plt.show()

prob = (duration < 9).sum() / sims
print(f"Estimated probability of Duration < 9: {prob:.4f}")
```
