```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  
import seaborn as sns
from scipy.stats import norm, binom
import scipy.stats as stats

simulations = 1000000  
coin_flips = np.random.randint(0, 2, (simulations, 10))  # 0 = tails, 1 = heads
num_heads = coin_flips.sum(axis=1)  

plt.figure(figsize=(6, 3))
plt.hist(num_heads, bins=range(12), density=True, alpha=0.6, color='b', align='left')
plt.axvline(8, color='r', linestyle='dashed', linewidth=2)  
plt.title("Number of Heads in 10 Coin Tosses")
plt.xlabel("Number of Heads")
plt.ylabel("Probability")
plt.xticks(range(11))
plt.show()

probability_8_heads = (num_heads == 8).sum() / simulations
print(f"Estimated probability of getting exactly 8 heads out of 10 tosses: {probability_8_heads:.4f}")
```
