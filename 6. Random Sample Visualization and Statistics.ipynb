```python
#gona generate a sample of 100,000 random data using np.random.seed(),
#then i visualize it with a histogram, and analyze its statistical properties

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis
```

```python
# seed and generate data
np.random.seed(42)
random_data = np.random.normal(loc=50, scale=10, size=100000)

#Visual hist
plt.figure(figsize=(10, 6))
sns.histplot(random_data, bins=50, kde=True, color='red')
plt.title("Histogram of 100,000 Random Data Points (Normal Distribution)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

# Simple statistical analysis
mean_value = np.mean(random_data)
median_value = np.median(random_data)
std_deviation = np.std(random_data)

print("=== Statistical Summary of the Data ===")
print(f"Mean: {mean_value:.2f}")
print(f"Median: {median_value:.2f}")
print(f"Standard Deviation: {std_deviation:.2f}")
```

```python
# Conclusion 
#- the data follows a normal distribution.
#- it shows a symmetric spread, with mean and median being very close.
# This type of data is suitable for many statistical analyses such as regression, t-tests, etc.
```
