```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr

# Linear + relation
np.random.seed(0)
x_linear = np.arange(1, 21)
y_linear = 2 * x_linear + np.random.normal(0, 2, size=len(x_linear))

# No clear relationship (random data)
x_random = np.arange(1, 21)
y_random = np.random.uniform(10, 40, size=len(x_random))

#visual
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.scatter(x_linear, y_linear, color='purple')
plt.title("Dataset 1: Strong Linear Relationship")
plt.xlabel("X Data 1")
plt.ylabel("Y Data 1")
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.scatter(x_random, y_random, color='cyan')
plt.title("Dataset 2: Random Data (No Clear Relationship)")
plt.xlabel("X Data 2")
plt.ylabel("Y Data 2")
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Correlation calculation
pearson1, _ = pearsonr(x_linear, y_linear)
spearman1, _ = spearmanr(x_linear, y_linear)

pearson2, _ = pearsonr(x_random, y_random)
spearman2, _ = spearmanr(x_random, y_random)

print("Correlation Results for Dataset 1")
print(f"Pearson Correlation: {pearson1:.4f}")
print(f"Spearman Correlation: {spearman1:.4f}")

print("\nCorrelation Results for Dataset 2")
print(f"Pearson Correlation: {pearson2:.4f}")
print(f"Spearman Correlation: {spearman2:.4f}")
```
