```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  
import seaborn as sns
from scipy import stats  
```

```python
data = {
    "Height": [170, 165, 180, 175, 160, 172, 168, 182, 177, 169],
    "Weight": [70, 65, 80, 75, 60, 72, 68, 85, 78, 66]
}

# Create DataFrame
df = pd.DataFrame(data)
print("DataFrame:\n", df)
```

```python
average_height = df["Height"].mean()
average_weight = df["Weight"].mean()

print(f"Average Height: {average_height:.2f} cm")
print(f"Average Weight: {average_weight:.2f} kg")
```

```python
counts, bin_edges = np.histogram(df["Height"], bins=5)  
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2 

plt.figure(figsize=(6,5)) 
plt.plot(bin_centers, counts, marker='o', linestyle='-', color='red', linewidth=2, label="Frequency Line")
plt.hist(df["Height"], bins=5, color='purple', edgecolor='black')
plt.xlabel("Height (cm)")
plt.ylabel("Frequency")
plt.title("Height Distribution Histogram")
plt.legend()
plt.show()
```

```python
counts, bin_edges = np.histogram(df["Weight"], bins=5)  
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2 

plt.figure(figsize=(10,5)) 
plt.plot(bin_centers, counts, marker='o', linestyle='-', color='red', linewidth=2, label="Frequency Line")
plt.hist(df["Weight"], bins=6, color='purple', edgecolor='black')
plt.xlabel("Weight (kg)")
plt.ylabel("Frequency")
plt.title("Weight Distribution Histogram")
plt.legend()
plt.show()
```
