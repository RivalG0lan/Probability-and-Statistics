```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis
```

```python
sample_data = [1, 2, 2, 3, 4, 5, 5, 5, 6]

#measures of Central Tendency

mean_value = np.mean(sample_data)
median_value = np.median(sample_data)
mode_value = pd.Series(sample_data).mode()[0]

print("=== Measures of Central Tendency ===")
print(f"Mean: {mean_value:.2f} (The arithmetic average of all values)")
print(f"Median: {median_value:.2f} (The middle value in sorted data)")
print(f"Mode: {mode_value:.2f} (The most common value in the dataset)")
```

```python
# Measures of Spread (Variability)
variance_value = np.var(sample_data, ddof=1)
std_dev_value = np.std(sample_data, ddof=1)
range_value = np.max(sample_data) - np.min(sample_data)
iqr_value = np.percentile(sample_data, 75) - np.percentile(sample_data, 25)

print("\n=== Measures of Spread ===")
print(f"Variance: {variance_value:.2f} (Average squared deviation from the mean)")
print(f"Standard Deviation: {std_dev_value:.2f} (Square root of the variance)")
print(f"Range: {range_value:.2f} (Max value - Min value)")
print(f"IQR: {iqr_value:.2f} (Interquartile Range - spread of middle 50%)")
```

```python
skewness_value = skew(sample_data)
kurtosis_value = kurtosis(sample_data)

print("\n=== Shape of the Data ===")
print(f"Skewness: {skewness_value:.2f} (Positive skew: right tail, Negative skew: left tail)")
print(f"Kurtosis: {kurtosis_value:.2f} (High kurtosis: heavy tails, Low kurtosis: light tails)")
```

```python
# Histogram with KDE
plt.figure(figsize=(12, 5))
sns.histplot(sample_data, bins=30, kde=True, color='blue')
plt.title("Histogram with KDE (Distribution of Data)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
# BOXPLOT
```

```python
# Boxplot
plt.figure(figsize=(8, 4))
sns.boxplot(x=sample_data, color='orange')
plt.title("Boxplot of Data (Visualizing Spread & Outliers)")
plt.xlabel("Value")
plt.show()
```

```python
# Pie Chart
categories = ['A', 'B', 'C', 'D']
values_pie = [25, 35, 20, 20]

plt.figure(figsize=(6, 6))
plt.pie(values_pie, labels=categories, autopct='%1.1f%%', colors=['red', 'blue', 'green', 'purple'])
plt.title("Pie Chart Example")
plt.show()
```

```python
# Bar Chart
plt.figure(figsize=(6, 4))
sns.barplot(x=categories, y=values_pie, palette='pastel')
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
# Line Chart
time = np.arange(1, 11)
line_values = np.random.randint(10, 100, size=10)

plt.figure(figsize=(6, 4))
plt.plot(time, line_values, marker='o', linestyle='-', color='b')
plt.title("Line Chart Example")
plt.xlabel("Time")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```

```python
# Scatter Plot
np.random.seed(42)
x_vals = np.random.rand(100)
y_vals = 3 * x_vals + np.random.normal(0, 0.2, 100)  # Linear relationship with noise

plt.figure(figsize=(6, 6))
sns.scatterplot(x=x_vals, y=y_vals, color='red')
plt.title("Scatter Plot (Relationship Between X and Y)")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
