```python
import numpy as np
np.random.seed(42)
data = np.random.normal(50, 10, 100)  # Sample data
mean_sample = np.mean(data)
variance_sample = np.var(data, ddof=1)

print("=== Point Estimation ===")
print(f"Sample Mean Estimate: {mean_sample:.2f}")
print(f"Sample Variance Estimate: {variance_sample:.2f}")
```

```python
import numpy as np
import scipy.stats as stats

np.random.seed(42)
data = np.random.normal(50, 10, 100)  # Sample data
mean_sample = np.mean(data)
std_sample = np.std(data, ddof=1)  # Sample standard deviation
n = len(data)
confidence = 0.95

# Compute confidence interval
t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)
margin_error = t_critical * (std_sample / np.sqrt(n))
ci_lower = mean_sample - margin_error
ci_upper = mean_sample + margin_error

print("=== Confidence Interval ===")
print(f"Mean Estimate: {mean_sample:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
import numpy as np
import scipy.stats as stats

np.random.seed(42)
data = np.random.normal(50, 10, 100)  # Sample data

# Hypothesis testing parameters
pop_mean = 50
alpha = 0.05

# One-sample t-test
t_stat, p_value = stats.ttest_1samp(data, pop_mean)

print("\n=== Hypothesis Testing (One-Sample t-test) ===")
print(f"Test Statistic: {t_stat:.2f}")
print(f"P-Value: {p_value:.4f}")

if p_value < alpha:
    print("Reject H0: There is significant evidence that the mean is different from 50.")
else:
    print("Fail to Reject H0: There is no significant evidence that the mean is different from 50.")
```

```python
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
group1 = np.random.normal(52, 10, 100)  # Sample data for Group 1
group2 = np.random.normal(48, 10, 100)  # Sample data for Group 2

alpha = 0.05  # Significance level

# Two-Sample t-test (Independent)
t_stat2, p_value2 = stats.ttest_ind(group1, group2)

print("\n=== Two-Sample t-test ===")
print(f"Test Statistic: {t_stat2:.2f}")
print(f"P-Value: {p_value2:.4f}")

if p_value2 < alpha:
    print("Reject H0: The two groups have significantly different means.")
else:
    print("Fail to Reject H0: The two groups do not have significantly different means.")

plt.figure(figsize=(8, 5))
sns.histplot(group1, color='blue', label='Group 1', kde=True, alpha=0.6)
sns.histplot(group2, color='red', label='Group 2', kde=True, alpha=0.6)
plt.title("Comparison of Two Groups")
plt.legend()
plt.show()
```
