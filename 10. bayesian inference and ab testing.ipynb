```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import norm

# Define x range from 0 to 1
x = np.linspace(0, 1, 100)

# Prior belief: uniform distribution Beta(1,1)
prior = stats.beta(1, 1).pdf(x)

# Posterior belief after seeing 7 heads in 10 coin flips
posterior = stats.beta(8, 4).pdf(x)  # 7 heads + 1 prior = 8, 3 tails + 1 prior = 4

# Plot prior and posterior
plt.plot(x, prior, label='Prior Beta(1,1)')
plt.plot(x, posterior, label='Posterior Beta(8,4)')
plt.legend()
plt.title("Prior vs Posterior (Beta Distribution)")
plt.xlabel("Probability of Heads")
plt.ylabel("Density")
plt.grid(True, alpha=0.3)
plt.show()

# A/B Testing
success_A = 30
trials_A = 100

success_B = 45
trials_B = 100

# Proportions
prop_A = success_A / trials_A
prop_B = success_B / trials_B

# Pooled proportion
pooled_prop = (success_A + success_B) / (trials_A + trials_B)

# Standard error
se = np.sqrt(pooled_prop * (1 - pooled_prop) * (1/trials_A + 1/trials_B))

# Z-score
z_score = (prop_A - prop_B) / se

# P-value (two-tailed)
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Output
print(f"Proportion A: {prop_A:.2f}, Proportion B: {prop_B:.2f}")
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("There is a significant difference between A and B.")
else:
    print("There is no significant difference between A and B.")
```
