```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import skew, kurtosis, ttest_1samp
```

```python
# Load dataset from Kaggle
df = pd.read_csv(....\example data set\netflix_titles.csv")

# Extract numeric values from duration column
df['duration_int'] = df['duration'].str.extract(r'(\d+)').astype(float)

# Separate data into movies and TV shows
movies = df[df['type'] == 'Movie'].copy()
tv_shows = df[df['type'] == 'TV Show'].copy()

# Descriptive statistics for movies
print("=== Descriptive Statistics for Movies ===")
print("Mean:", movies['duration_int'].mean())
print("Median:", movies['duration_int'].median())
print("Mode:", movies['duration_int'].mode()[0])

# Descriptive statistics for TV shows
print("\n=== Descriptive Statistics for TV Shows ===")
print("Mean:", tv_shows['duration_int'].mean())
print("Median:", tv_shows['duration_int'].median())
print("Mode:", tv_shows['duration_int'].mode()[0])

# Histogram visualization
sns.set(style="whitegrid")
fig, axs = plt.subplots(1, 2, figsize=(14, 5))

# Movie histogram
sns.histplot(movies['duration_int'], bins=30, kde=True, ax=axs[0], color="skyblue")
axs[0].set_title('Movie Duration Distribution (Minutes)')
axs[0].set_xlabel('Duration (minutes)')
axs[0].set_ylabel('Number of Movies')

# TV show histogram
sns.histplot(tv_shows['duration_int'], bins=10, kde=False, ax=axs[1], color="salmon")
axs[1].set_title('TV Show Season Count Distribution')
axs[1].set_xlabel('Number of Seasons')
axs[1].set_ylabel('Number of TV Shows')

plt.tight_layout()
plt.show()
```

```python
# Monte Carlo Simulation - 1000 samples of size 100 from movie durations
simulated_means = []
for _ in range(1000):
    sample = movies['duration_int'].dropna().sample(n=100, replace=True)
    simulated_means.append(sample.mean())

# Visualization of simulation results
plt.figure(figsize=(10, 5))
sns.histplot(simulated_means, bins=30, kde=True, color='purple')
plt.title('Monte Carlo Simulation: Average Movie Duration')
plt.xlabel('Average Duration (minutes)')
plt.ylabel('Frequency')
plt.axvline(np.mean(simulated_means), color='red', linestyle='--', label=f"Simulated Mean: {np.mean(simulated_means):.2f}")
plt.legend()
plt.show()
```

```python
# Skewness and kurtosis
movie_skew = skew(movies['duration_int'].dropna())
movie_kurtosis = kurtosis(movies['duration_int'].dropna())
print("=== Skewness & Kurtosis for Movies ===")
print("Skewness:", movie_skew)
print("Kurtosis:", movie_kurtosis)

tv_skew = skew(tv_shows['duration_int'].dropna())
tv_kurtosis = kurtosis(tv_shows['duration_int'].dropna())
print("\n=== Skewness & Kurtosis for TV Shows ===")
print("Skewness:", tv_skew)
print("Kurtosis:", tv_kurtosis)
```

```python
# Variability - Movies
movie_std = movies['duration_int'].std()
movie_var = movies['duration_int'].var()
movie_range = movies['duration_int'].max() - movies['duration_int'].min()

print("=== Variability of Movie Durations ===")
print("Standard Deviation:", movie_std)
print("Variance:", movie_var)
print("Range:", movie_range)

# Variability - TV Shows
tv_std = tv_shows['duration_int'].std()
tv_var = tv_shows['duration_int'].var()
tv_range = tv_shows['duration_int'].max() - tv_shows['duration_int'].min()

print("\n=== Variability of TV Show Seasons ===")
print("Standard Deviation:", tv_std)
print("Variance:", tv_var)
print("Range:", tv_range)
```

```python
# Inferential statistics - t-test (H0: mean = 100)
t_stat, p_val = ttest_1samp(movies['duration_int'].dropna(), 100)
print("=== One-Sample t-Test for Movie Duration (μ = 100) ===")
print("t-statistic:", t_stat)
print("p-value:", p_val)

if p_val < 0.05:
    print("Conclusion: Reject H0, the average movie duration is SIGNIFICANTLY different from 100 minutes.")
else:
    print("Conclusion: Fail to reject H0, the average movie duration is NOT significantly different from 100 minutes.")
```
